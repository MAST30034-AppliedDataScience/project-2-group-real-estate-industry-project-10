```python
import requests
import zipfile
import os
import io
import shutil

# Step 1: Download the ZIP file
url = 'https://s3.ap-southeast-2.amazonaws.com/cl-isd-prd-datashare-s3-delivery/Order_Y9LSRC.zip?orderid=FZ5O3I'
response = requests.get(url)

# Step 2: Create a temporary in-memory file and write the ZIP content to it
zip_file = zipfile.ZipFile(io.BytesIO(response.content))

# Step 3: Define the target directory where you want to save the files
save_directory = '../data/landing/External_data/Park/'

# Ensure the target directory exists
os.makedirs(save_directory, exist_ok=True)

# Step 4: Extract the ZIP file directly into the desired directory
zip_file.extractall(save_directory)

# Step 5: List extracted files to verify they were placed in the correct directory
extracted_files = os.listdir(save_directory)
print("Files extracted and saved to:", save_directory)
print("Extracted files:", extracted_files)
```
